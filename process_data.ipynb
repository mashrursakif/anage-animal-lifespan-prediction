```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import os
from pathlib import Path
from bs4 import BeautifulSoup
```

```python
data = pd.read_csv("an_age_data.csv")
data
```

```python
clean_data = data.copy()
clean_data[clean_data["Longevity"] == "Not yet established"] = np.nan
clean_data[clean_data["Typical body temperature"] == "Not yet available"] = np.nan
clean_data.dropna(subset=["HAGRID", "Species", "Name"], inplace=True)

cols_to_drop = [
    "Maximum longevity residual",
    "Maximum longevity",
    "Taxonomy",
    "Animal Diversity Web",
    "Unnamed: 42",
    "GenAge",
    "GenDR",
    "Internet",
    "Images",
    "Ageing Literature",
    "Integrated Taxonomic Information System",
    "Common name",
    "Source",
    "Sample size",
    "Data quality",
    "Observations",
    "Encyclopaedia of Life",
    "NCBI Taxonomy",
    "Entrez",
    "Synonyms",
]
clean_data.drop(columns=cols_to_drop, inplace=True)
```

```python
num_target_cols = [
    "Weight at hatching",
    "Clutch size",
    "IMR",
    "MRDT",
    "Resting metabolic rate",
    "Female sexual maturity",
    "Male sexual maturity",
    "Gestation",
    "Weaning",
    "Litter size",
    "Inter-litter interval",
    "Weight at birth",
    "Weight at weaning",
    "Adult weight",
    "Postnatal growth rate",
    "Basal metabolic rate",
    "Body mass",
    "Metabolic rate per body mass",
    "Incubation",
    "Metamorphosis",
    "Interbirth interval",
]

for col in num_target_cols:
    clean_data[col] = clean_data[col].str.replace(r"[^0-9.]", "", regex=True)
    clean_data[col] = clean_data[col].replace(r"^\s*$", np.nan, regex=True)
    clean_data[col] = clean_data[col].astype(float)
```

```python
# clean_data["Typical body temperature"] = clean_data["Typical body temperature"][0].split("º")[0]

clean_data["Typical body temperature"] = (
    clean_data["Typical body temperature"].str.split("º").str[0]
)
```

```python
clean_data["HAGRID"] = clean_data["HAGRID"].astype(int)
```

```python
clean_data[clean_data["Longevity"] == "--"] = np.nan
clean_data.dropna(subset=["Longevity"], inplace=True)
```

```python
clean_data.to_csv("an_age_data_clean.csv", index=False)
```
